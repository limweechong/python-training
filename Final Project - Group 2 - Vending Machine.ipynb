```python
# create a function for Vending Machine
#----------------------------------------------
def vending_machine():

    a = {'item': 'choc', 'price': 1.5, 'stock': 2}
    b = {'item': 'pop', 'price': 1.75, 'stock': 1}
    c = {'item': 'chips', 'price': 2.0, 'stock': 3}
    d = {'item': 'gum', 'price': 0.50, 'stock': 1}
    e = {'item': 'mints', 'price': 0.75, 'stock': 3}
    items = [a, b, c, d, e]
    value = 0 # value in vending machine

    print('Welcome to Group 2 Vending Machine! \n*******************************')

# Function to display all the items and prices
#----------------------------------------------
    def show(items):
        print('Items available \n*******************************')

        for item in items:      
            if item.get('stock') == 0:
                items.remove(item)
        for item in items:
            print(item.get('item'), item.get('price'))

        print('*******************************\n')

# Users selection and vending machine operation
#----------------------------------------------
    while True:
        show(items)
        selected = input('Select item: ')     #user selection assign to "selected"
        for item in items:                    #for loop to get item and retrieve the price
            if selected == item.get('item'):
                selected = item               
                price = selected.get('price')
                while value < price:          #while loop to prompt user to insert coin
                    value = float(input('Insert ' + str(price - value) + ': '))   
                else:                         #exit while loop and update the stock quantity
                    print('You have purchased: ' + selected.get('item'))
                    selected['stock'] -= 1
                    value -= price
                    print('Cash remaining: ' + str(value))
                    a = input('Buy something else? (y/n): ')
                    if a == 'n':
                        if value != 0:
                            print(str(value) + ' refunded')
                            value = 0
                            print('Thank you, have a nice day!\n')
                            #break 
                            return None
                        else:
                            print('Thank you, have a nice day!\n')
                            #break
                            return None                        
                    else:
                        continue
         

vending_machine()
```
